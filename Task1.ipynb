```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Data.csv')

# Plot histogram for years of insurance
plt.figure(figsize=(10, 6))
plt.hist(df['years_of_insurance_with_us'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Years of Insurance with the Company')
plt.xlabel('Years of Insurance')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
